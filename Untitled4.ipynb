```
import requests
import xmltodict
from bs4 import BeautifulSoup
from urllib.request import Request, urlopen
import pandas as pd
```

```
regnum_list=[];
noticeend_list=[];
age_list=[];
color_list=[];
kind_list=[];
neuteryn_list=[]
sex_list=[]
feature_list=[]
weight_list=[]
status_list=[]
```

```
# public api setting 

for i in range(1, 4):
  url="http://openapi.animal.go.kr/openapi/service/rest/abandonmentPublicSrvc/abandonmentPublic?serviceKey=Ymw9TAM%2F6p5qafOIRjCH2%2BPgeLNyNm4mwzowKDYJ9imPeOO%2FpJW6aGma%2BZs3W8tguq7gXODSBOq9dXiRscn0Tw%3D%3D&bgnde=20140601&endde=20200630&upkind=&kind=&upr_cd=&org_cd=&care_reg_no=&state=&pageNo="+str(i)+"&numOfRows=100&neuter_yn=&"
  resultxml=urlopen(url)
  result=resultxml.read()
  xmlsoup=BeautifulSoup(result, 'lxml-xml')
  items=xmlsoup.findAll("item")
  for item in items:
    regnum=item.find('desertionNo').text
    regnum_list.append(regnum)
    noticeend=item.find('noticeEdt').text
    noticeend_list.append(noticeend)
    age=item.find('age').text
    age_list.append(age)
    color=item.find('colorCd').text
    color_list.append(color)
    kind=item.find('kindCd').text
    kind_list.append(kind)
    neuteryn=item.find('neuterYn').text
    neuteryn_list.append(neuteryn)
    sex=item.find('sexCd').text
    sex_list.append(sex)
    feature=item.find('specialMark').text
    feature_list.append(feature)
    weight=item.find('weight').text
    weight_list.append(weight)
    status=item.find('processState').text
    status_list.append(status)
```

```
df=pd.DataFrame({'regnum':regnum_list, 'noticeend':noticeend_list, 'age':age_list, 'color':color_list,
                 'kind':kind_list, 'neuteryn':neuteryn_list, 'sex':sex_list, 'feature':feature_list, 'weight':weight_list, 'status':status_list})
df.head(30)
```
